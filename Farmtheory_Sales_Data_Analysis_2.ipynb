```python
import pandas as pd
```

```python
Buying_data=pd.read_csv("/content/Buying Data - Sheet3.csv")
Buying_data.head()
```

```python
Sales_data=pd.read_csv("/content/Sales Data1 - Sheet4.csv")
Sales_data.head()
```

```python
# Find the Sales Price for every SKU for the respective customer id on the given date

import pandas as pd
# Rename columns for consistency
Buying_data.rename(columns={'sku_name': 'SKU', 'Customer ID': 'CustomerID'}, inplace=True)
Sales_data.rename(columns={'sku_name': 'SKU', 'customer_id': 'CustomerID'}, inplace=True)

# Merge Buying_data and Sales_data on 'SKU', 'CustomerID', and 'Date'
merged_sales_data = pd.merge(Buying_data, Sales_data, on=['SKU'], how='left')

# Extract the desired columns
result_sales = merged_sales_data[['SKU', 'CustomerID', 'buying_Date', 'total_revenue']]

print(result_sales)
```

```python
# Find the total customer bill on the given date

total_bill_per_date = result_sales.groupby('buying_Date')['total_revenue'].sum()

print(total_bill_per_date)
```

```python
# Left Join Sales Data and Buying Data Tables (on sku_name and selling_date= (1+buying_date))
# Because today's buying SKU will be sold on next day

import pandas as pd
# Convert 'buying_Date' and 'selling_date' to datetime objects
Buying_data['buying_Date'] = pd.to_datetime(Buying_data['buying_Date'])
Sales_data['selling_date'] = pd.to_datetime(Sales_data['selling_date'])

# Calculate the next day after buying_date
Buying_data['next_day'] = Buying_data['buying_Date'] + pd.Timedelta(days=1)

# Perform the left join
merged_data = pd.merge(Sales_data, Buying_data,
                     left_on=['SKU', 'selling_date'],
                     right_on=['SKU', 'next_day'],
                     how='left')

print(merged_data)
```

```python
#  Calculate gross margin of given sku using BP and SP for the dataset

# Calculate gross margin for each row
merged_data['Gross Margin'] = 1 - (merged_data['buying_price'] / merged_data['total_revenue'])

# Display the result
print(merged_data[['SKU', 'buying_price', 'total_revenue', 'Gross Margin']])
```

```python
#  Find the highest selling gross margin SKU on respective dates

# Group by 'selling_date' and find the SKU with the highest 'Gross Margin' for each date
highest_margin_skus = merged_data.loc[merged_data.groupby('selling_date')['Gross Margin'].idxmax()]

# Display the result
print(highest_margin_skus[['selling_date', 'SKU', 'Gross Margin']])
```

```python
#  After joing Table, buying_price column would be NA or blank (as not every SKU is bought every day but might be sold particular day)
# Replace those NA or blank cells with the most recent date buying price of the given SKU

# Fill missing buying prices with the most recent price for each SKU
merged_data['buying_price'] = merged_data.groupby('SKU')['buying_price'].fillna(method='ffill')

print(merged_data)
```
